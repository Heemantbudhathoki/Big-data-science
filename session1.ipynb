```python
# challenge 1

# which airline has the highest number of delayed department ? = 

query1 = """SELECT a.name ,
                    COUNT(*) as num_delay FROM flights f
                    LEFT JOIN airlines a  ON f.carrier = a.carrier
                    WHERE f.dep_delay > 0
                    GROUP BY a.name
                    ORDER BY num_delay DESC
                    LIMIT 1"""
```

```python
result1 = spark.sql(query1)
result1.show()
```

```python
# challenge 2 : which airports has flights arriving most early in average

query2 = """SELECT ap.name,
                   AVG(f.arr_delay) AS avg_early
            FROM flights f
            LEFT JOIN airports ap
                 ON f.dest = ap.iata
            WHERE f.arr_delay < 0
            GROUP BY ap.name
            ORDER BY avg_early ASC
            LIMIT 1"""

result2 = spark.sql(query2)
result2.show()
```
